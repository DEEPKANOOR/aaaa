```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_check(df):
    x=df.drop('W',axis=1)
    vif=pd.DataFrame()
    vif['vif']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    vif['features']=x.columns
    print(vif)
```

```python
vif_check(df)
```

```python
# as ERA has the highest value of vif. m going to delete this.
```

```python
df=df.drop('ERA',axis=1)
vif_check(df)
```

```python
df=df.drop('H',axis=1)
vif_check(df)
```

```python
selected_feat=['E','R','SB','SV','ER','AB','SO']
```

```python
# now to find W, it's a regression problem.
```

```python
x_reg=pd.DataFrame()
for i in selected_feat:
    x_reg[i]=x[i]
x_reg
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,LassoCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.svm import SVR
lr=LinearRegression()
dtr=DecisionTreeRegressor()
rfr= RandomForestRegressor()
svr=SVR()
abr=AdaBoostRegressor()
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
def algo_check (x,y,algo):
    min_diff=1
    max_i=0
    for i in range(100):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = i)
        algo.fit(x_train,y_train)
        y_pred1 =algo.predict(x_train)
        acc1= r2_score(y_train,y_pred1)
        y_pred2 =algo.predict(x_test)
        acc2= r2_score(y_test,y_pred2)
        acc=acc1-acc2
        if acc< min_diff:
            min_diff=acc
            max_i = i
        i+=1
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = max_i)
    algo.fit(x_train,y_train)
    y_pred1 =algo.predict(x_train)
    acc1= r2_score(y_train,y_pred1)
    y_pred2 =algo.predict(x_test)
    acc2= r2_score(y_test,y_pred2)
    cvs=cross_val_score(algo,x_train,y_train,cv=5,scoring='r2')
    ac=cvs.mean()
    mae=mean_absolute_error(y_pred2,y_test)
    mse=mean_squared_error(y_pred2,y_test)
    rmse=np.sqrt(mse)
    print(f'''for algo {algo}, the training accuracy is {acc1}, tesing accuracy is {acc2},
    at random_state {max_i} and hence, mean square error is {mse} and mean_absolute_error 
    is {mae} and hence, rmse is {rmse}, also cross_validation_score is {ac}''') 
```

```python
algo_check(x_reg,y,lr)
```

```python
algo_check(x_reg,y,dtr)
```

```python
algo_check(x_reg,y,rfr)
```

```python
algo_check(x_reg,y,abr)
```

```python
algo_check(x_reg,y,svr)
```

```python
lasscv = LassoCV(alphas=[0.001,0.005,0.01,0.08,0.1,0.5,1,5,10],max_iter = 100, normalize =True)
algo_check(x_reg,y,lasscv)
alp=lasscv.alpha_
```

```python
lasso_reg =Lasso(alp)
algo_check(x_reg,y,lasso_reg)
```

```python
# from all of the above algos, it is clear that the cross validation score is the highest for Lasso model. also, the errors are not much higher for this algo..
```

```python
# hence, i m going to shortlist this algo Lasso at random_state 97.
```

```python
# hyperparameter tuning..
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_reg,y,test_size=0.25,random_state = 97)
lasso_reg.fit(x_train,y_train)
```

```python
params={'alpha':[0.001,0.005,0.01,0.02,0.05,0.1,0.5,1,1.5,5,10,12],'fit_intercept':[True,False],'copy_X':[True,False],'max_iter':[500,800,1000,1200],'positive':[True,False],'selection':['cyclic','random'],'warm_start':[True,False]}
grid=GridSearchCV(lasso_reg,param_grid=params,cv=7)
grid.fit(x_train,y_train)
print(grid.best_params_)
```

```python
lasso1=Lasso(alpha=1,copy_X=False,fit_intercept=True,max_iter=1200,positive=False,selection='random',warm_start=False)
lasso1.fit(x_train,y_train)
y_pred1 =lasso1.predict(x_train)
acc1= r2_score(y_train,y_pred1)
y_pred2 =lasso1.predict(x_test)
acc2= r2_score(y_test,y_pred2)
mae=mean_absolute_error(y_pred2,y_test)
mse=mean_squared_error(y_pred2,y_test)
rmse=np.sqrt(mse)
cvs=cross_val_score(lasso1,x_train,y_train,cv=5,scoring='r2')
ac=cvs.mean()
print(f'''for Lasso with grid search cv found parameters, the training accuracy is {acc1}, tesing accuracy is {acc2}
and hence, mean square error is {mse} and mean_absolute_error is {mae} and hence, rmse is {rmse}, , and cross validation score is {ac}''') 
```
